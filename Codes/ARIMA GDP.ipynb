```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
df=pd.read_excel(r"D:\Rosemeen Projects\World Economic Indicators\WorldBank.xlsx")
df.head()
```

```python
# Renaming Columns for better experience
df = df.rename(columns={'Birth rate, crude (per 1,000 people)':'BirthRate','Death rate, crude (per 1,000 people)':'DeathRate','Electric power consumption (kWh per capita)':'ElectricPowerConsumption','Infant mortality rate (per 1,000 live births)':'IMR','Unemployment (% of total labor force) (modeled ILO estimate)':'UnemploymentRate','Population density (people per sq. km of land area)': 'PopulationDensity','Life expectancy at birth (years)':'LifeExpectancyRate','Individuals using the Internet (% of population)':'InternetUsageRate'
})
```

```python
df.columns
```

```python
df = df.sort_values(["Country Name", "Year"])
```

```python
countries = df["Country Name"].unique()
countries
```

```python
results = []
for country in countries:
    country_df = df[df["Country Name"] == country]
    ts = country_df[["Year", "GDP per capita (USD)"]].copy()
    ts["Year"] = pd.to_datetime(ts["Year"], format="%Y")
    ts.set_index("Year", inplace=True)
    ts=ts.dropna()
    if len(ts)<10:
        continue
    train = ts.iloc[:-5]
    test = ts.iloc[-5:]
    try:
        model = ARIMA(train,order=(1,1,1))
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=5)
        mae = mean_absolute_error(
            test["GDP per capita (USD)"],
            forecast
        )
        results.append({
            "Country": country,
            "MAE": mae,
            "Forecast_Year_1": forecast.iloc[0],
            "Forecast_Year_2": forecast.iloc[1],
            "Forecast_Year_3": forecast.iloc[2],
            "Forecast_Year_4": forecast.iloc[3],
            "Forecast_Year_5": forecast.iloc[4]
        })
    except:
        continue
```

```python
results_df = pd.DataFrame(results)
results_df.sort_values("MAE", ascending=False).head()
```
